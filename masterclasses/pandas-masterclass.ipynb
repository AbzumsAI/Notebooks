## 16) Merge, Join & Concat

```python

customers = pd.DataFrame({
    "cust_id": [1, 2, 3],
    "name": ["Alice", "Bob", "Charlie"]
})
orders = pd.DataFrame({
    "order_id": [101, 102, 103, 104],
    "cust_id": [1, 2, 2, 4],
    "amount": [50, 20, 15, 99]
})

inner = customers.merge(orders, on="cust_id", how="inner")
left = customers.merge(orders, on="cust_id", how="left")
outer = customers.merge(orders, on="cust_id", how="outer", indicator=True)
print("Inner join:\n", inner)
print("\nLeft join:\n", left)
print("\nOuter join with indicator:\n", outer)
```

```python

# concat
east = pd.DataFrame({"region": ["EU", "EU"], "sales": [10, 20]})
west = pd.DataFrame({"region": ["US", "US"], "sales": [5, 15]})
concat_rows = pd.concat([east, west], axis=0, ignore_index=True)
concat_cols = pd.concat([east.reset_index(drop=True), west.reset_index(drop=True)], axis=1)
print(concat_rows)
print("\n---\n")
print(concat_cols)
```

## 19) Resampling & Rolling Windows

```python

# Make a longer daily series and resample to monthly
dates = pd.date_range("2024-01-01", periods=120, freq="D")
ts2 = pd.DataFrame({"value": rng.normal(100, 10, len(dates))}, index=dates)
monthly = ts2["value"].resample("M").mean()
monthly
```

```python

# Rolling mean and custom rolling apply
roll = ts2["value"].rolling(window=7, min_periods=3).mean()
custom = ts2["value"].rolling(window=5, min_periods=3).apply(lambda x: np.percentile(x, 90))
pd.DataFrame({"value": ts2["value"], "rolling_mean_7": roll, "rolling_p90_5": custom}).head(12)
```

```python

# Plot the monthly series and rolling mean (matplotlib, one chart per cell)
plt.figure()
plt.plot(monthly.index, monthly.values)
plt.title("Monthly Mean of Value")
plt.xlabel("Month")
plt.ylabel("Mean Value")
plt.show()
```

```python

plt.figure()
plt.plot(ts2.index, ts2["value"].values, label="Daily")
plt.plot(roll.index, roll.values, label="Rolling 7d Mean")
plt.title("Daily vs Rolling Mean")
plt.xlabel("Date")
plt.ylabel("Value")
plt.legend()
plt.show()
```

## 24) Reading in Chunks & Processing Large Files

## 25) Performance Tips & Timing

```python

# Vectorization vs apply
arr = rng.normal(0, 1, 1000000)
s = pd.Series(arr)

# timeit works best in Jupyter; simple manual timing here
import time
t0 = time.time()
v1 = (s * 2 + 1).sum()
t1 = time.time()
vec_time = t1 - t0

t0 = time.time()
v2 = s.apply(lambda x: x * 2 + 1).sum()
t1 = time.time()
apply_time = t1 - t0

{"vectorized_time_sec": vec_time, "apply_time_sec": apply_time, "same_result": np.isclose(v1, v2)}
```

```python

# Set dtypes early and use categoricals
df_perf = pd.DataFrame({
    "country": rng.choice(["DE", "FR", "US", "IT"], 10000),
    "city": rng.choice(["Berlin", "Paris", "NYC", "Rome"], 10000),
    "value": rng.normal(0, 1, 10000)
})
before = df_perf.memory_usage(deep=True).sum()
df_perf_opt = df_perf.astype({"country": "category", "city": "category", "value": "float32"})
after = df_perf_opt.memory_usage(deep=True).sum()
{"before_bytes": int(before), "after_bytes": int(after), "ratio": float(after/before)}
```

## 27) Real-World Mini-Project: Sales Analytics

```python

# Generate synthetic sales data
products = ["Widget", "Gadget", "Doohickey"]
regions = ["EU", "US", "APAC"]
sales = pd.DataFrame({
    "date": rng.choice(dates, 2000),
    "product": rng.choice(products, 2000),
    "region": rng.choice(regions, 2000),
    "units": rng.integers(1, 10, 2000),
    "unit_price": rng.choice([9.99, 12.5, 4.5], 2000)
})
sales["revenue"] = sales["units"] * sales["unit_price"]

sales.tail(20)
```

```python

# Top products by region
top = (sales.groupby(["region", "product"])["revenue"]
            .sum()
            .sort_values(ascending=False)
            .groupby(level=0)
            .head(3)
            .reset_index())
top
```

## 28) Real-World Mini-Project: A/B Test Analysis

```python

# Simulate A/B test clicks/conversions
n = 5000
ab = pd.DataFrame({
    "user_id": np.arange(n),
    "variant": rng.choice(["A", "B"], n, p=[0.5, 0.5]),
    "clicked": rng.choice([0,1], n, p=[0.7, 0.3])
})
ab.head(15)
```

```python

# Assume conversion happens only if clicked; B has a tiny lift
ab["converted"] = np.where(
    (ab["variant"]=="B") & (ab["clicked"]==1),
    rng.choice([0,1], size=n, p=[0.75, 0.25]),
    rng.choice([0,1], size=n, p=[0.8, 0.2])
)
ab.head(15)
```

```python

summary = ab.groupby("variant")[["clicked","converted"]].mean().rename(columns=lambda c: c+"_rate")
summary
```

```python

# Simple difference in conversion rates
diff = summary.loc["B","converted_rate"] - summary.loc["A","converted_rate"]
{"conversion_lift_B_minus_A": float(diff)}
```

## 29) Real-World Mini-Project: Parsing Web Logs

```python

# Create fake combined log lines
logs = pd.Series([
    '127.0.0.1 - - [10/Jan/2024:13:55:36 +0000] "GET /index.html HTTP/1.1" 200 1024',
    '127.0.0.1 - - [10/Jan/2024:13:56:01 +0000] "POST /login HTTP/1.1" 302 512',
    '10.0.0.5 - - [10/Jan/2024:14:00:12 +0000] "GET /dashboard HTTP/1.1" 500 256',
])
pattern = (r'(-P<ip>\S+) \S+ \S+ \[(-P<datetime>[^\]]+)\] '
           r'"(-P<method>\S+) (-P<path>\S+) \S+" '
           r'(-P<status>\d{3}) (-P<size>\d+)')
parsed = logs.str.extract(pattern)
parsed["status"] = parsed["status"].astype(int)
parsed["size"] = parsed["size"].astype(int)
parsed
```

```python

# Count by status and top paths
status_counts = parsed["status"].value_counts().sort_index()
top_paths = parsed["path"].value_counts().head(5)

print(status_counts)
print("\nTop paths:")
print(top_paths)
```

## 31) Tidy Data & Cleaning Example (End-to-End)

```python

raw = pd.DataFrame({
    "Full Name": [" alice  ", "BOB", "Charlie  "],
    "City": ["Berlin", "Paris", "Berlin"],
    "Age(yrs)": ["28", "thirty-four", None],
    "Has Car-": ["yes", "no", "yes"]
})
# Cleaning pipeline
def parse_age(x):
    try:
        return int(x)
    except:
        return np.nan

cleaned = (
    raw
    .rename(columns={"Full Name": "name", "Age(yrs)": "age", "Has Car-": "has_car"})
    .assign(name=lambda d: d["name"].str.strip().str.title(),
            age=lambda d: d["age"].apply(parse_age).astype("Float64"),
            has_car=lambda d: d["has_car"].str.lower().map({"yes": True, "no": False}))
    .astype({"City": "category"})
)

# Validate
assert cleaned["name"].notna().all()
cleaned
```

## 32) Common Gotchas: `SettingWithCopy`, chained indexing

```python

# Example: avoid chained indexing
df_g = pd.DataFrame({"x": [1,2,3], "y": [4,5,6]})
# Bad: df_g[df_g["x"]>1]["y"] = 0  # may raise a warning in real runs
# Good: use .loc with a mask
mask = df_g["x"] > 1
df_g.loc[mask, "y"] = 0
df_g
```

## 34) Cheat Sheet & Further Reading

- Core objects: `Series`, `DataFrame`, `Index`
- Selection: `loc`, `iloc`, boolean masks; avoid chained indexing
- Aggregation: `groupby` + `agg`/`transform`/`filter`
- Reshaping: `melt`, `pivot`, `pivot_table`, `stack`/`unstack`
- Merge data: `merge`, `join`, `concat`
- Time series: `to_datetime`, set index, `resample`, `rolling`
- Text: `.str` methods, regex with `extract`/`replace`
- Missing data: `isna`, `fillna`, `interpolate`
- Dtypes: prefer nullable (`Int64`, `boolean`, `string`), `category` for repeated labels
- Performance: vectorize; set dtypes early; use chunked reads for large files
- IO: `read_csv`, `read_json`, `read_excel`, `to_parquet` (if available)

**Official docs:** https://pandas.pydata.org/docs/
