
## 2. `ndarray` Basics <a id="ndarray-basics"></a>

The core of NumPy is the **ndarray**, a homogeneous N-dimensional array optimized for speed and vectorized operations.


````python

# Creating a simple array from a Python list
a = np.array([1, 2, 3, 4])
print("a:", a)
print("type:", type(a))
print("dtype:", a.dtype)
print("ndim:", a.ndim)
print("shape:", a.shape)
print("size:", a.size)
print("itemsize (bytes per element):", a.itemsize)
````


## 4. Data Types (`dtype`) <a id="data-types-dtype"></a>

Arrays are homogeneous: every element has the same type. Choosing the right `dtype` affects memory and precision.


````python

ints = np.array([1, 2, 3], dtype=np.int32)
floats = np.array([1, 2, 3], dtype=np.float64)
complexes = np.array([1+2j, 3+4j], dtype=np.complex128)
print(ints.dtype, floats.dtype, complexes.dtype)
print("memory (bytes):", ints.nbytes, floats.nbytes, complexes.nbytes)
````

````python

# Casting and safe conversions
x = np.array([1.2, 2.5, 3.7], dtype=np.float32)
print("original:", x, x.dtype)
print("astype int:", x.astype(np.int32))
print("astype complex:", x.astype(np.complex128))
````


## 8. Broadcasting <a id="broadcasting"></a>

Broadcasting lets you combine arrays of different shapes by virtually expanding dimensions (no data copy).
**Rule of thumb:** Align shapes from the right; dimensions must match or be 1.


````python

A = np.arange(6).reshape(2,3)
b = np.array([10, 20, 30])        # shape (3,)
print("A:\n", A)
print("A + b:\n", A + b)         # add row-wise
print("A * b:\n", A * b)
````

````python

# Column-wise broadcasting by adding a new axis
c = np.array([100, 200]).reshape(2,1)  # shape (2,1)
print("A + c:\n", A + c)
````


**Visualization of shapes** (right-aligned):

```
A: (2, 3)
b:    (3)  -> treated as (1, 3)
c: (2, 1)
```



## 9. Universal Functions (ufuncs) & Vectorization <a id="ufuncs-vectorization"></a>


## 10. Reductions, Aggregations, and Axes <a id="reductions-aggregations-axes"></a>

Aggregations have an `axis` argument. `axis=0` reduces rows (down columns), `axis=1` reduces columns (across rows).


````python

X = np.arange(1,13).reshape(3,4)
print("X:\n", X)
print("sum all:", X.sum())
print("sum axis=0 (col sums):", X.sum(axis=0))
print("sum axis=1 (row sums):", X.sum(axis=1))
print("mean, std:", X.mean(), X.std())
print("min axis=1:", X.min(axis=1))
print("argmax axis=0:", X.argmax(axis=0))
````


## 18. Memory Layout, Views vs Copies, and Strides <a id="memory-views-strides"></a>

Slicing usually returns **views** (no copy). Fancy indexing returns **copies**. Understand `strides` and memory order (C vs Fortran).


````python

A = np.arange(12).reshape(3,4)
sub = A[::2, ::2]      # view
print("sub:", sub)
fancy = A[[0,2], [0,3]]  # copy
sub[0,0] = -1
print("A after modifying view:\n", A)
print("fancy (independent copy):", fancy)
print("C-contiguous:", A.flags['C_CONTIGUOUS'], "F-contiguous:", A.flags['F_CONTIGUOUS'])
````

````python

# Demonstrate Fortran order
B = np.asfortranarray(A.copy())
print("B (Fortran): F-contiguous-", B.flags['F_CONTIGUOUS'])
````


## 19. Performance: Vectorization and Timing <a id="performance-vectorization-timing"></a>

Vectorization avoids Python loops. Use `timeit` to compare. (In Jupyter, you can also use `%%timeit` magic.)


````python

import timeit

def python_loop(n=100_000):
    s = 0.0
    for i in range(n):
        s += i*i
    return s

def numpy_vectorized(n=100_000):
    x = np.arange(n, dtype=np.float64)
    return np.dot(x, x)

print("python_loop:", timeit.timeit("python_loop()", globals=globals(), number=5))
print("numpy_vectorized:", timeit.timeit("numpy_vectorized()", globals=globals(), number=5))
````


## 21. Images as Arrays: Mini Project <a id="images-as-arrays"></a>

We'll synthesize a grayscale image (2D array) and apply simple transformations (thresholding, blur via convolution).


````python
# Create a synthetic image: concentric circles
n = 256
yy, xx = np.mgrid[:n, :n]
center = (n//2, n//2)
radius = np.sqrt((yy-center[0])**2 + (xx-center[1])**2)

print("yy:\n", yy)
print("xx:\n", xx)
print("center:\n", center)
````

````python

# Create a synthetic image: concentric circles
n = 256
yy, xx = np.mgrid[:n, :n]
center = (n//2, n//2)
radius = np.sqrt((yy-center[0])**2 + (xx-center[1])**2)
img = np.exp(-(radius**2)/(2*(n/6)**2))
plt.figure()
plt.imshow(img)
plt.title("Synthetic Image (grayscale)")
plt.axis('off')
plt.show()
````

````python

# Thresholding
th = 0.5
binary = (img > th).astype(float)
plt.figure()
plt.imshow(binary)
plt.title("Thresholded")
plt.axis('off')
plt.show()
````

````python

# Simple blur via 2D convolution with a Gaussian-like kernel
def conv2d(image, kernel):
    kh, kw = kernel.shape
    ph, pw = kh//2, kw//2
    # pad
    padded = np.pad(image, ((ph,ph),(pw,pw)), mode='edge')
    out = np.empty_like(image)
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            window = padded[i:i+kh, j:j+kw]
            out[i,j] = np.sum(window * kernel)
    return out

# Build a separable Gaussian kernel
def gaussian_kernel(size=7, sigma=1.5):
    ax = np.arange(-(size//2), size//2+1)
    g = np.exp(-(ax**2)/(2*sigma**2))
    g = g / g.sum()
    kern = np.outer(g, g)
    return kern

kernel = gaussian_kernel(9, 2.0)
blurred = conv2d(img, kernel)
plt.figure()
plt.imshow(blurred)
plt.title("Blurred Image")
plt.axis('off')
plt.show()
````


## 23. Fast Fourier Transform (FFT) Basics <a id="fft-basics"></a>

Use `np.fft` for frequency-domain analysis (e.g., denoising, signal processing).


````python

# Make a noisy signal: two sines + noise
rng = np.random.default_rng(0)
t = np.linspace(0, 1, 1024, endpoint=False)
sig = np.sin(2*np.pi*50*t) + 0.5*np.sin(2*np.pi*120*t) + 0.2*rng.normal(size=t.size)

# FFT
F = np.fft.rfft(sig)
freqs = np.fft.rfftfreq(t.size, d=(t[1]-t[0]))

plt.figure()
plt.plot(freqs, np.abs(F))
plt.title("Magnitude Spectrum")
plt.xlabel("Frequency (Hz)"); plt.ylabel("|F|")
plt.show()
````


## 27. Custom ufuncs with `vectorize` and `frompyfunc` <a id="custom-ufuncs"></a>

Wrap Python functions for convenience (not speed). Prefer true vectorization when possible.


````python

def is_prime_py(n: int) -> bool:
    if n < 2: return False
    if n % 2 == 0 and n != 2: return False
    r = int(math.sqrt(n)) + 1
    for k in range(3, r, 2):
        if n % k == 0:
            return False
    return True

vec_is_prime = np.vectorize(is_prime_py, otypes=[bool])
nums = np.arange(1, 30)
print("primes mask:", vec_is_prime(nums))
````


## 28. Monte Carlo: Estimating π <a id="monte-carlo"></a>

Estimate π by sampling points in a unit square and counting those inside the unit circle.


````python

rng = np.random.default_rng(123)
N = 200_000
pts = rng.random((N, 2))
inside = (pts[:,0]**2 + pts[:,1]**2) <= 1.0
pi_hat = 4 * inside.mean()
print("pi estimate:", pi_hat)
````

````python

# Visualize a small subset
sub = pts[:1000]
mask = (sub[:,0]**2 + sub[:,1]**2) <= 1.0
plt.figure()
plt.scatter(sub[~mask,0], sub[~mask,1], s=5)
plt.scatter(sub[mask,0], sub[mask,1], s=5)
plt.title("Monte Carlo Points in Unit Square")
plt.xlabel("x"); plt.ylabel("y")
plt.show()
````


## 29. Precision Pitfalls & Best Practices <a id="precision-pitfalls"></a>

Floating point is finite precision. Beware of catastrophic cancellation and order of operations.


````python

# Catastrophic cancellation example
a = 10_000.0
b = 9_999.999
print("Direct difference:", a - b)
# Kahan summation demo
x = np.full(1_000_000, 1e-6)
print("naive sum:", x.sum())
def kahan_sum(arr):
    s = 0.0
    c = 0.0
    for v in arr:
        y = v - c
        t = s + y
        c = (t - s) - y
        s = t
    return s
print("kahan sum:", kahan_sum(x))
````
