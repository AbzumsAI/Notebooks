# Session 4, Part 1: Functions

Cell 1: Defining a Function:

```python
# Basic function that returns a greeting
def greet(name):
    return f"Hello, {name}!"

# Call the function
greet("Dr. Smith")
```

Cell 2: Parameters & Arguments

```python
# Function with two parameters
def bmi(weight, height):
    return weight / (height ** 2)

# Call with arguments
bmi(70, 1.75)
```

Cell 3: Default Parameters

```python
# Function with a default parameter
def greet(name, title="Mr./Ms."):
    return f"Hello, {title} {name}!"

print(greet("Jones"))
print(greet("Adams", "Dr."))
```

Cell 4: Return Values and Data Types

```python
def calculate_stats(values):
    avg = sum(values) / len(values)
    return avg, max(values), min(values)

calculate_stats([72, 88, 90, 65, 78])  # Returns a tuple
```

Cell 5: Local Variable

```python
def say_hello():
    name = "Alice"  # Local variable
    print("Hello", name)

say_hello()
# print(name)  #  This would cause a NameError
```

Cell 6: Global Variable

```python
age = 20  # Global variable

def greet():
    # global age
    age = 19
    print("You are", age)

greet()
print("Outside function:", age)
```

Cell 7:  Local Variable Shadows Global

```python
x = 10  # Global

def show():
    x = 5  # Local, shadows global
    print("Inside function:", x)

show()
print("Outside function:", x)
```

Cell 8: Creating Global from Inside Function

```python
def define_global():
    global city
    city = "Tehran"
    print(city)

define_global()
city = 'London'
print("Global variable created:", city)
```

# Session 4, Part 2: Lambda Functions

Cell 1: Introduction to Lambda

```python
# Regular function vs lambda
def add(x, y):
    return x + y

add_lambda = lambda x, y: x + y

# Call both
add(3, 5), add_lambda(3, 5)
```

Cell 2: Syntax of Lambda

```python
# Syntax: lambda arguments: expression
triple = lambda x: x * 3
triple(4)
```

Cell 3: Use Case: `sorted()` with Lambda

```python
# Sorting a list of tuples by second value (age)
patients = [("Alice", 34), ("Bob", 52), ("Zara", 28)]
sorted_patients = sorted(patients, key=lambda x: x[1])
sorted_patients
```

Cell 4: Use Case: `map()` with Lambda

```python
# Convert temperatures from Celsius to Fahrenheit
temps_c = [36.6, 37.0, 38.5]
temps_f = list(map(lambda c: round((c * 9/5) + 32, 1), temps_c))
temps_f
```

Cell 5: Use Case: `filter()` with Lambda

```python
# Filter high blood pressures (>140)
bp_readings = [120, 130, 150, 160, 138]
high_bp = list(filter(lambda bp: bp > 140, bp_readings))
high_bp
```

Cell 14: Raising Your Own Exceptions

```python
def check_age(age):
    if age < 0:
        raise ValueError("Age cannot be negative")
    return f"Your age is {age}"

check_age(-3)
```

```python
def quick_cvd_risk(age, bp, smoker):
    score = 0
    if age > 50: score += 2
    if bp > 140: score += 2
    if smoker: score += 3

    if score >= 5:
        return "High risk"
    elif score >= 3:
        return "Moderate risk"
    else:
        return "Low risk"
    

print(quick_cvd_risk(51, 150, False))
```
