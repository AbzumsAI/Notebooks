# Session 3, Part 1: Control Flow

Cell 1: Basic if statement

```python
x = 7

if x > 5:
    print("x is greater than 5")
```

Cell 2: if and else

```python
x = 5

if x % 2 == 0:
    print("Even number")
else:
    print("Odd number")
```

Cell 3: if, elif, and else

```python
x = 0

if x > 0:
    print("Positive")
elif x == 0:
    print("Zero")
else:
    print("Negative")
```

Cell 4: Nested if statements

```python
x = 10

if x > 0:
    if x % 2 == 0:
        print("Positive even number")
    else:
        print("Positive odd number")
```

Cell 5: Multiple elif conditions

```python
grade = 85

if grade >= 90:
    print("A")
elif grade >= 80:
    print("B")
elif grade >= 70:
    print("C")
else:
    print("F")
```

# Session 3, Part 2: `for` Loops

Cell 1: Basic for loop over a list

```python
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print("I like", fruit)
```

Cell 2: for loop with `range()`

```python
for i in range(5):
    print(f"I Luv Python for {i} days.")
```

Cell 3: for loop with start, stop, step in range()

```python
for i in range(9, 0, -2):
    print("Odd number:", i)


for i in range(1, 10, 2):
    print("Odd number:", i)
```

Cell 4: Loop over a string

```python
word = "python"

for letter in word:
    print(letter)
```

Cell 5: Looping over a list with `enumerate()`

```python
colors = ["red", "green", "blue"]

for index, color in enumerate(colors):
    print(f"Color at index {index} is {color}")
```

Cell 6: Looping over a dictionary

```python
person = {"name": "Ali", "age": 22, "city": "Tehran"}

for key, value in person.items():
    print(f"{key}: {value}")
```

Cell 7: Nested for loops

```python
for i in range(1, 4): # i = 1, 2, 3
    for j in range(1, 6): # j = 1, 2, 3, 4, 5
        print(f"{i} x {j} = {i*j}")
```

# Session 3, Part 3: `while` Loops

Cell 1: Basic loop with a condition

```python
radius = int(input("Enter primary radius:"))

while 3.14 * radius ** 2 >= 1:
    print(f"Radius is {radius}, Area is { 3.14 * radius ** 2}.")
    radius /= 2
```

Cell 2: infinite loop

```python
sum = 0

while True:
    in_ = int(input("Enter a number:"))
    if in_ == -1:
        break
    sum += in_
    print("SUM:", sum)

print("TOTAL SUM:", sum)
```

Cell 3: continue

```python
while True:
    height, weight = input("Enter height (m) and weight (kg)").split()
    height, weight = float(height), float(weight)
    if height < 0.3 or weight < 20:
        print("Unable to calculate BMI.")
        continue
    print("BMI is: ", weight / (height **2))
```
