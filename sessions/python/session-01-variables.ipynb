# Session 2, Part 3: Data Types

Cell 1: Basic data types

```python
a = 10         # int
b = 3.14       # float
c = "Python"   # str
d = True       # bool

print(type(a), type(b), type(c), type(d))
```

Cell 2: Type conversion (casting)

```python
x = 5.7
y = int(x)   # float to int
z = str(x)   # float to str

print("x:", x, "| y:", y, "| z:", z)
print(type(y), type(z))
```

```python
s = True

y = int(s)  # bool to int
z = str(s)  # bool to str
print("y:", y, "| z:", z)
print(type(y), type(z))
```

```python
t = 0

y = float(t)  # int to float
z = str(t)    # int to str
w = bool(t)  # int to bool
print("y:", y, "| z:", z, "| w:", w)
print(type(y), type(z), type(w))
```

Cell 3: String to number conversion

```python
s = "-2"
num = int(s)
print("Converted:", num, "| Type:", type(num))
```

Cell 4: Using type() and isinstance()

```python
value = 3.14

print(type(value))                 # returns <class 'float'>
print(isinstance(value, float))   # returns True
print(isinstance(value, int))     # returns False
```

Cell 5: Special data type None

```python
x = None
print(x)
print(type(x))
```

Cell 6: Checking multiple types

```python
data = [1, "hello", 3.5, False]

for item in data:
    print(f"{item} is of type {type(item)}")
```

Cell 7: List

```python
capitals = ['Tehran', 'Paris', 'Tokyo', 'Berlin', 'London', 'Moscow', 'Beijing', 'New Delhi']

print("List of capitals:", capitals)
print("First capital:", capitals[0])
print("Last capital:", capitals[-1])
print("Slice of capitals:", capitals[1:3])
print("Capitals in reverse:", capitals[::-1])
print("Capitals in sorted order:", sorted(capitals))
print("Capitals in original order:", capitals)
print("Length of capitals list:", len(capitals))

capitals.append('Cairo')
print("After appending Cairo:", capitals)
capitals.insert(2, 'Madrid')
print("After inserting Madrid at index 2:", capitals)
capitals.remove('Berlin')
print("After removing Berlin:", capitals)
last = capitals.pop()
print("After popping last element:", capitals)
print("Popped element:", last)
```

# Session 2, Part 4: Strings

Cell 1: Define and print a string

```python
message = "Hello, Python!"
message2 = 'Hello, Java!'
message3 = '''Hello, Kotlin!
This is a multi-line string.'''
print(message)
print(message2)
print(message3)
```

Cell 2: String indexing and slicing

```python
text = "Programming"

print("First character:", text[0])
print("Last character:", text[-1])
print("First 5 characters:", text[:5])
print("From 4th to end:", text[3:])
```

Cell 3: String concatenation and repetition

```python
greet = "Hi"
name = "Ali"

print(greet + " " + name)
print("Python " * 3)
```

Cell 4: String length and membership

```python
msg = "Machine Learning"

print("Length:", len(msg))
print("Learn" in msg)
print("Deep" not in msg)
```

Cell 5: Changing case

```python
s = "hello world! this is a python course."

print(s.upper())
print(s.lower())
print(s.title())
print(s.capitalize())
```

Cell 6: Removing spaces and special characters

```python
raw = "   data science   \n"
print("Before:", raw)
print("Before:", repr(raw))
print("After strip():", repr(raw.strip()))
```

Cell 7: Splitting and joining strings

```python
sentence = "AI is&the future"
words = sentence.split('&')
print("Split into list:", words)

joined = "-".join(words)
print("Joined with '-':", joined)
```

Cell 8: Replace, find, and count

```python
quote = "data is the new oil"

print(quote.replace("data", "AI"))
print("Position of 'new':", quote.find("new"))
print("Count of 'the':", quote.count("the"))
```

Cell 9: f-strings for formatting

```python
language = "Python"
version = 3.10

print(f"{language} version {version} is awesome!")
```

# Session 2, Part 5: Operators

Cell 1: Arithmetic Operators

```python
a = 10
b = 3

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Exponentiation:", a ** b)
print("Square root:", a ** 0.5)
```

Cell 2: Assignment Operators

```python
x = 5
x += 2  # x = x + 2
print("x += 2:", x)

x -= 1
print("x -= 1:", x)

x *= 3
print("x *= 3:", x)

x /= 2
print("x /= 2:", x)
```

Cell 3: Comparison Operators

```python
a = 10
b = 5

print("a == b:", a == b)
print("a != b:", a != b)
print("a > b:", a > b)
print("a < b:", a < b)
print("a >= b:", a >= b)
print("a <= b:", a <= b)
```

Cell 4: Logical Operators

```python
x = True
y = False

print("x and y:", x and y)
print("x or y:", x or y)
print("not x:", not x)
```

Cell 5: Identity Operators

```python
a = [1, 2, 3]
b = a
b.append(4)
print("a:", a)
print("b:", b)

c = [1, 2, 3]

print("a is b:", a is b)
print("a is c:", a is c)
print("a is not c:", a is not c)
```

Cell 6: Membership Operators

```python
nums = [1, 2, 3, 4, 5]
print("3 in nums:", 3 in nums)
print("6 not in nums:", 6 not in nums)

students = {'John': 85, 'Alice': 90, 'Bob': 78}
print("'John' in students:", 'John' in students)
print("85 in students:", 85 in students.values())
```

Cell 7: Bitwise Operators

```python
a = 5       # 0b0101
b = 3       # 0b0011

print("a & b:", a & b)   # Bitwise AND
print("a | b:", a | b)   # Bitwise OR
print("a ^ b:", a ^ b)   # Bitwise XOR
print("~a:", ~a)         # Bitwise NOT
print("a << 1:", a << 1) # Left Shift
print("a >> 1:", a >> 1) # Right Shift
```
