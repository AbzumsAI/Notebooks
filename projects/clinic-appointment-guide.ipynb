# Python Project Guide

## Project: Clinic Appointment

### Purpose

Create a simple command-line Python program that helps manage patient information in a small clinic. Students will practice using Python basics by registering patients, booking appointments, tracking symptoms, and generating simple reports.


---

## Step 1: Plan Your Data

1. **Patients**: Use a dictionary to store patient information.
2. **Available Slots**: Create a list of free time slots (e.g., "Monday 9:00").
3. **Appointments**: Use a dictionary to link patients to their booked time.
4. **Symptoms**: Prepare a dictionary to map each symptom to a severity level.

**Goal:** Know what data you need and how to organize it.

---

## Step 2: Build a Menu

1. Show a list of choices:

   * 1: Add Patient
   * 2: Book Appointment
   * 3: Update Symptoms
   * 4: Show Patient Report
   * 5: Exit
2. Use a `while` loop to keep the program running.
3. Ask the user to pick an option.
4. Use `if`/`else` to do the right action for each choice.

**Goal:** Make a menu that repeats until the user chooses to exit.

---

## Step 3: Add a Patient

1. Ask the user for:

   * Name, Age, Gender
   * Weight, Height
   * Conditions (comma-separated)
   * Symptoms (comma-separated)
2. Clean the input (e.g., split commas and remove spaces).
3. Store all of it in the patients dictionary.

**Goal:** Save patient information in a clean and structured way.

---

## Step 4: Book an Appointment

1. Show the list of available slots.
2. Ask for the patient's name and slot number.
3. Check:

   * Is the patient already registered-
   * Is the slot available-
   * Has the patient already booked-
4. If valid, assign the slot and remove it from the list.

**Goal:** Let patients book free slots and prevent duplicates.

---

## Step 5: Update Symptoms

1. Ask for the patient's name.
2. Get a list of new symptoms.
3. Add them to the existing symptoms.
4. Use `lambda` and `map()` to turn symptoms into severity scores.
5. Use `filter()` to find symptoms with high severity (e.g., over 3).
6. Show all symptoms and the critical ones.

**Goal:** Update and analyze symptoms using Python tools.

---

## Step 6: Show Patient Report

1. Print patient information: name, age, gender, weight, height.
2. Show BMI (calculated from weight and height).
3. List all symptoms and highlight severe ones.
4. Show booked appointment if it exists.
5. Calculate a simple risk score (based on age, BMI, symptoms).

**Goal:** Show a summary of each patient's health status.

---

## Step 7: Handle Errors

1. Catch wrong input (e.g., letters instead of numbers).
2. Handle cases like:

   * Patient not found
   * Invalid slot number
   * Empty input
3. Show helpful error messages.

**Goal:** Prevent your program from crashing with bad input.

---


## Python Skills Used

* **Variables**: Store patient and appointment data
* **Data Types**: Strings, numbers, lists, dictionaries
* **Operators**: Calculate BMI and scores
* **If/Else**: Check conditions and make choices
* **Loops**: Repeat the menu and go through lists
* **Functions**: Handle tasks like BMI and reports
* **Lambda, Map, Filter**: Analyze symptoms
* **Try/Except**: Handle invalid input

---

### Final Tips

* Build the project one step at a time.
* Test each part before adding the next.
* Keep the code clean and readable.
* Ask questions and work with others to improve your ideas.
